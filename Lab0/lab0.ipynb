```python
import random

random_values = [random.uniform(0,500) for _ in range(100)]

print(random_values)
```

```python
import numpy as np

random_values_np = np.array(random_values)

mean = np.mean(random_values_np)
std_biased = np.std(random_values_np, ddof=0)
std = np.std(random_values_np, ddof=1)

mean,std_biased,std
```

```python
import numpy as np

samples = np.random.uniform(0,500,100)
avg = samples.mean()
std = samples.std(ddof=1)

print("The average is:", avg)
print("The standard deviation is:", std)
```
